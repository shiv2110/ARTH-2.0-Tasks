```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
face_cascade = cv2.CascadeClassifier("./haarcascade_xml_files/haarcascade_frontalface_default.xml")
img1 = cv2.imread("./assets/musk.jpg")
img2 = cv2.imread("./assets/hykes.png")
```

```python
print(img1.shape, img2.shape)
```

```python
img1 = cv2.resize(img1, (300, 300))
```

```python
plt.imshow(img1)
plt.axis(False)
```

```python
plt.imshow(img2)
plt.axis(False)
```

```python
gray1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
face = face_cascade.detectMultiScale(gray1, 1.3, 3)

for (x, y, w, h) in face:
    cv2.rectangle(img1, (x, y), (x + w, y + h), (255, 0, 0), 1)
cv2.imshow('Elon Musk', img1)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
face = face_cascade.detectMultiScale(gray2, 1.3, 3)

for (x1, y1, w1, h1) in face:
    cv2.rectangle(img2, (x1, y1), (x1 + w1, y1 + h1), (255, 0, 0), 1)
cv2.imshow('Solomon Hykes', img2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
musk_face = img1[y:(y + h), x:(x + w)]
```

```python
plt.imshow(musk_face)
plt.axis(False)
```

```python
hykes_face = img2[y1:(y1 + h1), x1:(x1 + w1)]
plt.imshow(hykes_face)
plt.axis(False)
```

```python
print(musk_face.shape, hykes_face.shape)
```

```python
musk_face1 = cv2.resize(musk_face, hykes_face.shape[:2])
hykes_face1 = cv2.resize(hykes_face, musk_face.shape[:2])
```

```python
print(musk_face1.shape, hykes_face1.shape)
```

```python
img1[y:(y + h), x:(x + w)] = hykes_face1
img2[y1:(y1 + h1), x1:(x1 + w1)] = musk_face1
```
